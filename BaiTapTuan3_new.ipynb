Viết một chương trình nhập vào một số nguyên n và in ra các chữ số đảo ngược nhau. 


```python
n=int(input())
def dn(n):
    cs=0
    while n>=10:
        cs=cs*10+n%10
        n=n//10     
    cs=cs*10+n
    return cs
print(dn(n))
```

Kiểm tra xem một số có phải là lũy thừa của 2 hay không
Sử dụng các toán tử bitwise để xác định xem một số có phải là lũy thừa của 2 hay không.
Gợi ý: Đối với lũy thừa của 2, n & (n-1) == 0

```python
n=int(input())
#code 1
def kt(n):
    return n>0 and (n&(n-1)==0)
#ki hieu & là and trong nhị phân
#code 2
def np(n):
    if n == 1:
        return True
    if n % 2 != 0 or n <= 0:
        return False
    return np(n // 2)
#print(np(n))
print(kt(n))
```

- 0-50 kWh: 1500đ/kWh
- 51-100 kWh: 2000đ/kWh
- 100 kWh: 3000đ/kWh

```python
n=int(input())
t=n
s=0
if t>100:
    s=s+(t-100)*3000
    t=100
if t>50:
    s+=(t-50)*2000
    t=50
if t>0:
    s=s+t*1500
print(s)
```
