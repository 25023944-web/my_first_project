 Nhập vào một số nguyên dương n, kiểm tra xem n có phải là số dạng 2^k hay không 

```python
n=int(input())
def kt(n):
    while n%2==0:
        n=n//2
    if n==1:return True
    return False
if kt(n):print("Yes")
else:print("No")
```

Nhập vào A, tìm n nhỏ nhất sao cho
1 + 1/2 + 1/3 + 1/4 + ... + 1/n > A 

```python
n=int(input())
k=0
s=0
while s<n:
    k+=1
    s+=1/k
print(k-1)
 
```

Nhập vào một dãy số nguyên, ngưng nhập khi người dùng nhập -1.
Sau khi nhập xong, in số lớn nhất, số nhỏ nhất trong những số vừa nhập

```python
k=-9999999999
p=9999999999
while True:
    n=int(input())
    k=max(n,k)
    p=min(n,k)
    if n==-1:break
print(k,p)
```

Dãy số fibonacci là dãy số được định nghĩa như sau: 1, 1, 2, 3, 5, 8, 13,... với số kế tiếp sẽ
bằng tổng hai số trước đó
Nhập vào số nguyên dương A, hãy tìm số trong dãy số fibonacci lớn nhất nhưng không vượt
quá A 


```python
n=int(input())
f1=1
f2=1
while f2<=n:
    f1,f2=f2,f1+f2
print(f1)
```

Nhập vào một chuỗi, hãy đếm xem trong chuỗi có bao nhiêu từ (quy định là chuỗi không có
ký tự đặc biệt, không số, không có dấu câu, chỉ có ký tự chữ và khoảng trắng)
Dùng cách: for trên index; KHÔNG dùng hàm split(), len()

```python
n=input()
d=0
s=0
l=0
for char in n:
    l+=1
for i in range(0,l):
    if n[i]==" ":
        d+=1
    else:d=0
    if d==1:s+=1 
if n[0]==" ":s-=1
if n[l-1]==" ":s=s-1
print(s+1)
```

```python
n=input()
l=0
vt=0
s=""
n=" "+n
for char in n:
    l+=1
for i in range(0,l):
    if n[i]!=" ":
        vt=i
    if vt!=0:
        s=s+n[i]
        if n[i]==" ":break
print(s)
```

Nhập vào một chuỗi có dạng 3 số nguyên, mỗi số nguyên cách nhau một dấu phẩy, hãy tính
tổng 3 số nguyên đó
VD:
Nhập: 3, 12, 15
Tổng: 30 


```python
n=input()
s=" "
a=[]
t=0
for i in n:
    if i!=",":
        s=s+i
    else:
        a.append(int(s))
        s=" "
a.append(int(s))
for i in a:
    t=t+i
print(t)
```

Nhập vào một chuỗi, hãy đếm xem trong chuỗi có bao nhiêu ký tự in hoa, bao nhiêu ký tự in
thường, bao nhiêu ký tự số 

```python
n=input()
d1,d2,d=0,0,0
for i in n:
    if "a"<=i<="z":d+=1
    if "A"<=i<="Z":d1+=1
    if "0"<=i<="9":d2+=1
print(d,"chu thuong",d1,"chu hoa",d2,"chu so")
```
