Các quy tắc cần cài đặt: 
1. Giá mở cửa (0–0.8 km): 
- Xe 4 chỗ: 11,000 VND 
- Xe 7 chỗ: 12,000 VND 
2. Giá theo km sau 0.8 km: 
- Xe 4 chỗ: từ 0.8→20 km ơnh 13,000 VND/km; trên 20 km tính 10,500 VND/km 
- Xe 7 chỗ: từ 0.8→20 km ơnh 14,000 VND/km; trên 20 km tính 12,000 VND/km

chờ). 
5. Điều kiện đặc biệt: 
- Nếu km > 30 và không phải ban đêm, áp dụng giảm 10% trên tổng tiền. 
- Nếu tổng tiền cuối cùng < 50,000 nhưng km > 5, đặt tổng tiền bằng 50,000 VND (mức tối thiểu)

```python
lx,km,pc,pt=map(int,input().split())
def cp(lx,km,pc,pt):
    s=0
    p=km
    if lx==4:
        s=11000
        if km>20:
            s=s+10500*(km-20)
            km=20
        if km>0.8:
            s=s+13000*(km-0.8)
        if pc>5:
            s=s+(pc//5)*2000
        if pt in [22,23,0,1,2,3,4,5]:s+=s*0.2
        elif p>30:s=s*0.9
        if s<50000 and p>5:s=50000
    else:
        s=12000
        if km>20:
            s=s+12000*(km-20)
            km=20
        if km>0.8:
            s=s+14000*(km-0.8)
        if pc>5:
            s=s+(pc//5)*2000
        if pt in [22,23,0,1,2,3,4,5]:s+=s*0.2
        elif p>30:s=s*0.9
        if s<50000 and p>5:s=50000
    return s
print(cp(lx,km,pc,pt))
```

Yêu cầu: Viết hàm detect_number_type(s) trả về 'int' nếu s biểu diễn một số nguyên, 'float' nếu biểu diễn một số thực dấu 
chấm động, ngược lại trả về 'neither'. 
- Đầu vào: chuỗi s. 
- Đầu ra: một trong các giá trị {'int', 'float', 'neither'}. 
- Quy tắc: 
- Cho phép có dấu + hoặc - ở đầu. 
- Số nguyên phải chỉ gồm các chữ số và không có dấu chấm thập phân (ví dụ: '100'). 
- Số thực phải có dấu chấm và có chữ số ở cả hai phía của dấu chấm (ví dụ: '100.0', '-3.14'). 
- Các chuỗi như '.5', '5.', '1e3' được xem là 'neither' trong bài này.

```python
def dx(s):
    t=s
    if t.startswith('+') or t.startswith('-'):
        t=t[1:]
    if not t:
        return 'neither'
    if t.isdigit():
        return 'int'
    if '.' in t:
        p=t.split('.')
        if len(p)==2 and p[0].isdigit() and p[1].isdigit():
            return 'float'
    return 'neither'
s=input()
print(dx(s))
```

Viết hàm trả về tam giác vuông cân cao n bằng ký tự ''.
Dòng thứ i (đánh số từ 1) có i dấu '', các dòng nối với nhau bằng ký tự xuống dòng.
Ví dụ với n = 4 


```python
n=int(input())
def kt(n):
    for i in range(1,n+1):
        for j in range(1,i+1):
            print("*",end="")
        print(end="\n")
kt(n)
```

Viết hàm trả về chuỗi biểu diễn hình chữ nhật N x M bằng ký tự '', các hàng cách nhau bởi ký tự xuống dòng.

```python
n,m=map(int,input().split())
def kt(n,m):
    for i in range(0,n):
        for j in range(0,m):
            print("*",end="")
        print(end="\n")
kt(n,m)
```
