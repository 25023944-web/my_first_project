Hàm nguyên tố

```python
n = int(input())
a = list(map(int, input().split()))
def nto(n):
    if n<2:return False
    if (n==2) or (n==3) or (n==5) or (n==7):
        return True
    if (n%2==0) or (n%3==0) or (n%5==0) or (n%7==0):
        return False
    i=5
    while (i*i<=n) :
        i=i+6
        if (i%n==0) or (i%(n+2)==0):return False
    return True
for i in range(len(a)):
    if nto(a[i]) :
        print(a[i],end="")
```
