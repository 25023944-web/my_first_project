```python
def tinh_tien_dien():
    # Nhập dữ liệu
    ten_chu_ho = input("Ten chu ho: ")
    chi_so_cu = int(input("Chi so thang truoc: "))
    chi_so_moi = int(input("Chi so thang nay: "))
    
    # Tính lượng điện tiêu thụ
    usage = chi_so_moi - chi_so_cu
    tong_tien_truoc_thue = 0
    temp_usage = usage

    # Tính toán theo từng bậc (lũy tiến)
    if temp_usage > 400:
        tong_tien_truoc_thue += (temp_usage - 400) * 3460
        temp_usage = 400
    if temp_usage > 300:
        tong_tien_truoc_thue += (temp_usage - 300) * 3350
        temp_usage = 300
    if temp_usage > 200:
        tong_tien_truoc_thue += (temp_usage - 200) * 2998
        temp_usage = 200
    if temp_usage > 100:
        tong_tien_truoc_thue += (temp_usage - 100) * 2380
        temp_usage = 100
    if temp_usage > 50:
        tong_tien_truoc_thue += (temp_usage - 50) * 2050
        temp_usage = 50
    if temp_usage > 0:
        tong_tien_truoc_thue += temp_usage * 1984

    # Tính VAT 8% và ép kiểu về số nguyên (lấy đến đơn vị đồng)
    tong_phai_tra = int(tong_tien_truoc_thue * 1.08)
    
    # Xuất kết quả
    print(f"Ho va ten: {ten_chu_ho}")
    print(f"Tien phai tra la: {tong_phai_tra}")

tinh_tien_dien()
```
