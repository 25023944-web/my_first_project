 Giải hệ phương trình tuyến tính:
ax + by = m,
cx + dy = n

```python
a,b,m,c,d,n=map(int,input().split())
def ucln(a,b):
    while b!=0:
        a,b=b,a%b
    return abs(a)
k=ucln(a,c)
b=b*(c/k)
m=m*(c/k)
d=d*(a/k)
n=n*(a/k)
print(round((m-b*((m-n)/(b-d)))/(k*a)),round((m-n)/(b-d)))
```
