Hàm nguyên tố <Tốt>

```python
n=int(input())
def nto(n):
    if n<2:return False
    if n in [2,3,5,7]:return True
    if (n%2==0) or (n%3==0) or (n%5==0) or (n%7==0):return False
    i=5
    while i*i<=n:
        i+=6
        if (n%i==0) or (n%(i+2)==0):return False
    return True
for i in range(1000):
    if (nto(i)):print(i)
```

Tìm ước chung lớn nhất và bội chung nhỏ nhất

```python
a,b=map(int,input().split())
def ucln(a,b):
    while b!=0:
        a,b=b,a%b
    return abs(a)
print(ucln(a,b))
print(a*b/ucln(a,b))
```

Sàng nguyên tố

```python
n=int(input())
if n<2: 
    print("0")
else:
    a=[1]*(n+1)
    a[0]=0
    a[1]=0
    for i in range(2,int(n**0.5+1)):
        if a[i]==1:
            for j in range(i*2,n+1,i):
                a[j]=0
for i in range(1,n+1):
    if a[i]==1:print(i)
```
