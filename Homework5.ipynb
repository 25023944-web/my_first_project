Viết hàm kiểm tra một số có phải số Amstronhg hay không? Hàm trả về kiểu bool, có tham số là một số 
nguyên x

```python
n=int(input())
def am(n):
    cs=0
    k=len(str(n))
    p=n
    while n>0:
        cs=cs+(n%10)**k
        n=n//10
    if p==cs:return True
    return False
print(am(n))
```

Viết hàm detect_number_type(s) trả về 'int' nếu s là số nguyên, 'float' nếu là số thực, ngược lại 'neither'.

```python
s=input().strip()
def detect(s):
    if not s:return'neither'
    n,start,dot,num=len(s),0,0,0
    if s[0]in'-+':
        start=1
        if n==1:return'neither'
    for i in range(start,n):
        c=s[i]
        if c=='.':
            dot+=1
            if dot>1:return'neither'
        elif'0'<=c<='9':num+=1
        else:return'neither'
    if num==0:return'neither'
    return'float'if dot==1 else'int'
print(detect(s))
```

Viết hàm point_in_quadrilateral(quad, p) kiểm tra điểm p có nằm trong hoặc trên biên của tứ giác quad hay 
không. Trong đó: quad là một tứ giác (4 đỉnh – 4 điểm), p là một điểm cần kiểm tra. VD: quad = [(0,0), (4,0), (4,3), 
(0,3)], p = (2,1)

```python
def tich(A, B, P):
 return (B[0] - A[0]) * (P[1] - A[1]) - (B[1] - A[1]) * (P[0] - A[0])
def kt(nhap, p):
    A = nhap[0]
    B = nhap[1]
    C = nhap[2]
    D = nhap[3]
    c1 = tich(A, B, p)
    c2 = tich(B, C, p)
    c3 = tich(C, D, p)
    c4 = tich(D, A, p)
    a = (c1 < 0) or (c2 < 0) or (c3 < 0) or (c4 < 0)
    b = (c1 > 0) or (c2 > 0) or (c3 > 0) or (c4 > 0)
    if a and b:return False
    else:return True 
quad = []
for i in range(4):
    x,y=map(float,input().split())
    quad.append((x, y))
px ,py=map( float,input().split())
p = (px, py)
kq = kt(quad,p)
print(kq)
```

Viết hàm tính góc ∠ABC (độ) từ tọa độ A(x1,y1), B(x2,y2), C(x3,y3)

```python
import math
def tt(x1, y1, x2, y2, x3, y3):
    tu = (x1-x2)*(x3-x2) + (y1-y2)*(y3-y2)
    mau = (((x1-x2)**2 + (y1-y2)**2)**0.5) * (((x3-x2)**2 + (y3-y2)**2)**0.5)
    if mau == 0: return 0 
    d = max(-1.0, min(1.0, tu/mau))
    radian = math.acos(d)
    return math.degrees(radian)
x1, y1, x2, y2, x3, y3=map(int,input().split())
kq= tt(x1, y1, x2, y2, x3, y3)
print(kq)
```

Viết hàm sum_1_to_n(n) trả về tổng các số x từ 1 đến n, nếu x có các chữ số đôi một khác nhau
Nếu n ≤ 0 trả về 0

```python
n=int(input())
def kt(n):
    a=[0]*10
    while n>0:
        a[n%10]+=1
        n=n//10
    for i in a:
        if i>=2:return False
    return True
def tong(n):
    s=0
    for i in range(1,n+1):
        if kt(i):s+=i
    return s
print(tong(n))
```
