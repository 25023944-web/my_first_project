Viết chương trình in ra hình tam giác cân có độ cao h được nhập từ bàn phím 

```python
n=int(input())
d=2
k=1+(n-1)*2
c=k//2+1
for i in range(1,k+1):
    if i==c:
        print("*")
        break
    else:print(end=" ")
while d<n:
    t=k//2+d
    for i in range(0,k):
        if (i>0):
            if (i==t-1) or (i==k-t+1):print("*",end=" "),
            else:print(end=" ")
    d+=1
    print(end="\n")
#print(" ",end="")
for i in range(0,k):
    print("*",end="")
```

Nhập vào số nguyên dương a, kiểm tra xem a có phải là số hoàn hảo hay không.
Số hoàn hảo là số có tổng các ước (ước nhỏ hơn nó) bằng chính nó, VD: số 6, 28… 

```python
n=int(input())
def kt(n):
    s=0
    for i in range(1,n):
        if n%i==0:
            s=s+i
    if s==n:return True
    return False
if kt(n):print("Yes")
else:print("No")
```

4: Nhập vào một dãy số nguyên, dừng nhập khi người dùng nhập -1. Sau khi nhập
xong, in số lớn nhất, số nhỏ nhất trong những số vừa nhập. KHÔNG ĐƯỢC dùng mảng

```python
def nhap(a,b):
    n=int(input())
    if n==-1:
        print(a,b)
        return a,b
    else :
        a=max(a,n)
        b=min(b,n)
        return nhap(a,b)
k=nhap(-10000000,10000000)
```

Dãy số fibonacci là dãy số được định nghĩa như sau: 1, 1, 2, 3, 5, 8, 13,... với số
kế tiếp sẽ bằng tổng hai số trước đó. Nhập vào số nguyên dương A, hãy tìm số trong
dãy số fibonacci lớn nhất nhưng không vượt quá A 

```python
n=int(input())
f1=1
f2=1
while f1+f2<n:
    f1,f2=f2,f1+f2
print(f2)
```

```python
n=int(input())
def nto(n):
    if n<2 :return False
    if n in [2,3,5,7]:return True
    if (n%2==0) or (n%3==0) or (n%5==0) or (n%7==0):return False
    i=5
    while (i*i<=n):
        i+=6
        if (n%i==0) or (n%(i+2)==0):return False
    return True
def kt(n):
    d=0
    while n!=0:
        if (nto(n%10)):d+=1
        n=n//10
    return d
print(kt(n))
```
